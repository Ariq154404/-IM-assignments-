```python
import numpy as np
import math
import matplotlib.pyplot as plt
n = np.arange(0., 7., 0.01)
k=4
plt.plot(n,np.log(n),label='logarithmic')
plt.plot(n,n,label='linear')
plt.plot(n,n**2,label="quadratic")
plt.plot(n,n**3,label="cubic")
plt.plot(n,n**k,label="polynomial,k=4")
plt.xlabel("n")
plt.ylabel("O(f(n))")
plt.legend()
plt.show()
```

It can be observed from the above graphs, when being plotted against the vaues of n that the complexity of the functions increases from logarithmic to potynomial and peaks at polynomial with k=4.
